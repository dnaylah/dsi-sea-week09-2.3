```python
import pandas as pd
import numpy as np
```

```python
#load all into csv's
sunrun = pd.read_csv("run.csv")
ideal_power = pd.read_csv("ipwr.csv")
ballard_power = pd.read_csv("bldp.csv")
synthesis = pd.read_csv("symx.csv")
ge = pd.read_csv("ge.csv")
```

```python
#deal with - 
sunrun.loc[2,'Low'] = sunrun['Low'].mean()
synthesis.loc[0,"Open"] = 0
synthesis.loc[0,"Low"] = 0
synthesis.loc[0,"Volume"] = 0
synthesis[['Open','High','Low','Close','Volume']] = synthesis[['Open','High','Low','Close','Volume']].apply(pd.to_numeric)

synthesis.loc[0,'Open'] = synthesis['Open'].mean()
synthesis.loc[0,'Low'] = synthesis['Low'].mean()
synthesis.loc[0,'Volume'] = synthesis['Volume'].mean()
```

```python
#convert everything to numeric except the dates
sunrun[['Open','High','Low','Close','Volume']] = sunrun[['Open','High','Low','Close','Volume']].apply(pd.to_numeric)
ideal_power[['Open','High','Low','Close','Volume']] = ideal_power[['Open','High','Low','Close','Volume']].apply(pd.to_numeric)
ballard_power[['Open','High','Low','Close','Volume']] = ballard_power[['Open','High','Low','Close','Volume']].apply(pd.to_numeric)
synthesis[['Open','High','Low','Close','Volume']] = synthesis[['Open','High','Low','Close','Volume']].apply(pd.to_numeric)
ge[['Open','High','Low','Close','Volume']] = ge[['Open','High','Low','Close','Volume']].apply(pd.to_numeric)
```

```python
# convert dates to datetype
sunrun['Date'] = pd.to_datetime(sunrun['Date'])
ideal_power['Date'] = pd.to_datetime(ideal_power['Date'])
ballard_power['Date'] = pd.to_datetime(ballard_power['Date'])
synthesis['Date'] = pd.to_datetime(synthesis['Date'])
ge['Date'] = pd.to_datetime(ge['Date'])
```

```python
# set all the indexes as the date
sunrun.set_index('Date', inplace=True)
ideal_power.set_index('Date', inplace=True)
ballard_power.set_index('Date', inplace=True)
synthesis.set_index('Date', inplace=True)
ge.set_index('Date', inplace=True)
```

```python
# add year and month columns
sunrun['Year'] = sunrun.index.year
sunrun['Month'] = sunrun.index.month
ideal_power['Year'] = ideal_power.index.year
ideal_power['Month'] = ideal_power.index.month
ballard_power['Year'] = ballard_power.index.year
ballard_power['Month'] = ballard_power.index.month
synthesis['Year'] = synthesis.index.year
synthesis['Month'] = synthesis.index.month
ge['Year'] = ge.index.year
ge['Month'] = ge.index.month
```

```python
# add day of week column
sunrun['DayOfWeek'] = sunrun.index.dayofweek
ideal_power['DayOfWeek'] = ideal_power.index.dayofweek
ballard_power['DayOfWeek'] = ballard_power.index.dayofweek
synthesis['DayOfWeek'] = synthesis.index.dayofweek
ge['DayOfWeek'] = ge.index.dayofweek
```

## Correleations of all the companies by day and mean

```python
def corre_by_day(comp,col):
    print "One day: ",comp[col].resample('D').mean().autocorr(lag=1)
    print "One month: ",comp[col].resample('D').mean().autocorr(lag=30)
    print "Two months: ",comp[col].resample('D').mean().autocorr(lag=60)
    print ""
```

```python
# cols has some pointless values, like days of week
cols2 = 'Open','High','Low','Close','Volume'
```

```python
for i,j in both:
    print "Company: ",j
    print ""
    for j in cols2:
        print "Column: ",j
        corre_by_day(i, j)
    print""
```

### companies with small variation by day, month, or 2 month, shows the general stability of the stock

## Correlations of all the companies by month, including median and mean

```python
def corre_by_month_median(comp,col):
    print ""
    print "One month median: ",comp[col].resample('M').median().autocorr(lag=1)
    
def corre_by_month_mean(comp,col):
    print "One month mean: ",comp[col].resample('M').mean().autocorr(lag=1)
    print ""
```

```python
for i,j in both:
    print "Company: ",j
    print ""
    for j in cols2:
        print "Column: ",j
        corre_by_month_median(i, j)
        corre_by_month_mean(i, j)
    print""
```

### Interesting find: the Volume category seems to better correlate with the median, whereas the other
### columns tend to better correlate with the mean
